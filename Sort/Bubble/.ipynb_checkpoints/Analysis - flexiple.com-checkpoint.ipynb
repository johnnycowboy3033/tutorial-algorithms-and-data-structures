```javascript
function printPointers(baseArr,baseIndex,plusOne){
    console.log('[index] = value');
    
        for( let index = 0; index < baseArr.length; index++){
        
        let label = "";
        
        if(baseIndex == index || plusOne == index){
            label = label + "<---- ";
            
            if(baseIndex == index){
                label = label +" BaseIndex";
            }
            
            if(plusOne == index){
                label = label +" PlusOne";
            }
            
        }
        
        
        
        console.log(`[ ${index} ] = ${baseArr[index]}  ${label} \n`)
    }
}
```

```javascript
let bubbleSort_ForLoop = (inputArr) => { 
    
    let len = inputArr.length;
    
    for (let i = 0; i < len; i++) {
        console.log('_____________________'); 
        for (let j = 0; j < len; j++) {
            
            if (inputArr[j] > inputArr[j + 1]) {
                
                let tmp = inputArr[j];
                inputArr[j] = inputArr[j + 1];
                inputArr[j + 1] = tmp;
            }
            printPointers(inputArr,i,j)
        }
    }
    
    return inputArr;
};
```

```javascript
bubbleSort_ForLoop ([6,4,2,5,7])
```

# Another Way

A slightly better solution involves tracking a variable called "checked" which is initially set to FALSE and becomes true when there is a swap during the iteration. Running this code on a do while loop to run the sorting function only when "checked" is true ensures that the function will not run on a sorted array more than once.

```javascript
let bubbleSort_WhileLoop = (inputArr) => {
    let len = inputArr.length;
    let checked;
    do {
        checked = false;
        console.log('_____________________'); 
        for (let i = 0; i < len; i++) {
            if (inputArr[i] > inputArr[i + 1]) {
                let tmp = inputArr[i];
                inputArr[i] = inputArr[i + 1];
                inputArr[i + 1] = tmp;
                checked = true;
            }
            printPointers(inputArr,i,i + 1)
        }
    } while (checked);
    return inputArr;
};
```

```javascript
bubbleSort_WhileLoop([6,4,2,5,7])
```
