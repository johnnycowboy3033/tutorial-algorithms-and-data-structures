```javascript
function printPointers(baseArr,baseIndex,PlusOne){
    console.log('[index] = value');
    
        for( let index = 0; index < baseArr.length; index++){
        
        let label = "";
        
        if(baseIndex == index || PlusOne == index){
            label = label + "<---- ";
            
            if(baseIndex == index){
                label = label +" BaseIndex";
            }
            
            if(PlusOne == index){
                label = label +" PlusOne";
            }
            
        }
        
        
        
        console.log(`[ ${index} ] = ${baseArr[index]}  ${label} \n`)
    }
}
```

```javascript
// Optimized javaScript implementation
// of Bubble sort
// An optimized version of Bubble Sort
function bubbleSort(arr, n)
{
    var i, j, temp;
    var swapped;
    for (i = 0; i < n - 1; i++) 
    {
        console.log('_____________________'); 
        swapped = false;
        for (j = 0; j < n - i -1; j++) 
        {
            if (arr[j] > arr[j + 1]) 
            {
                // Swap arr[j] and arr[j+1]
                temp = arr[j];
                arr[j] = arr[j + 1];
                arr[j + 1] = temp;
                swapped = true;
            }
            printPointers(arr,i,j);  
                      
        }
 
        // IF no two elements were 
        // swapped by inner loop, then break
        if (swapped == false)
        break;
    }
}
```
